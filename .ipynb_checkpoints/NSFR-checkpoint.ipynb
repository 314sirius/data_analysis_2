```python
import pandas as pd
import numpy as np

pd.options.display.float_format = '{:,.2f}'.format

balance_data = pd.read_csv('balance_data.csv')
nsfr_factor = pd.read_csv('nsfr_factor.csv')
```

```python
""" Use balance_data.csv and nsfr_factor.csv datasets to provide analytics on the capital and liquidity requirement of the bank.
    Provide a calculation of the NSFR ratio of the bank. 
"""
def task_4(balance_data, nsfr_factor):
    balance_data = balance_data.copy()
    nsfr_factor = nsfr_factor.copy()
    
    
    #join 2 tables
    merged=pd.concat([balance_data, nsfr_factor[['NSFR_FACTOR']]], axis=1)
    
    #calculate NSFR
    merged['Weighted']=merged['TOTAL_BALANCE_EUR']*merged['NSFR_FACTOR']
    NSFR=merged[merged['NSFR_GROUP']=='ASF']['Weighted'].sum()/merged[merged['NSFR_GROUP'].\
                                                                      isin(['RSF', 'RSF_OBS'])]['Weighted'].sum()
    #sort by gap in specific order: <6m, 6-12m, >12m
    custom_dict={'<6m':0, '6 - 12m':1, '>12m':2}
    merged=merged.sort_values(by='GAP', key=lambda col: col.map(custom_dict))
    
    #sort_values(), groupby(), agg() as sum(total_balance_eur), replace NaN cells with 0.00 eur, 
    #and .unstack() to change the view of the table. 
    
    balance_by_gap=merged.sort_values(['NSFR_GROUP','BALANCE_GROUP'], ascending=[True,True]).\
                    groupby(['NSFR_GROUP','BALANCE_GROUP','GAP'], sort=False).\
                    agg({'TOTAL_BALANCE_EUR':'sum'}).\
                    unstack().\
                    replace(np.nan, '0.00', regex=True)
           
    
    #calculate weighted average NSFR factors
    weighted_factor=merged.groupby(['NSFR_GROUP','BALANCE_GROUP']).agg({'Weighted':'sum', 'TOTAL_BALANCE_EUR':'sum'})
    weighted_factor['WEIGHTED_FACTORS']=weighted_factor['Weighted']/weighted_factor['TOTAL_BALANCE_EUR']
    weighted_factor=weighted_factor[['WEIGHTED_FACTORS']]
    
    # End of code segment
    
    print('\n4. Task solution')
    
    print('\nCurrent NSFR ratio is {:.2f}%'.format(NSFR*100))
        
    print('\nDaily balance by balance groups and time gaps:')
    display(balance_by_gap)
        
    print('\nWeighted NSFR haircuts by balance groups:')
    display(weighted_factor)

task_4(balance_data, nsfr_factor)
```

```python
import matplotlib.pyplot as plt
result=pd.concat([balance_data, nsfr_factor[['NSFR_FACTOR']]], axis=1)
result=result.sort_values('GAP').groupby(['BALANCE_GROUP', 'GAP'])['TOTAL_BALANCE_EUR'].sum().unstack()
result[['<6m', '6 - 12m', '>12m']].plot(kind='bar', stacked=True)

plt.title("Total balance by balance group")
plt.xlabel('Balance Group')
plt.ylabel('Total Balance, EUR')
plt.show()
```
